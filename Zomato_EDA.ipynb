```python
df_country = pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
# checking if country code is present in the first data set called zomato
df.columns
```

### combinig two data sets

```python
final_df = pd.merge(df, df_country, on='Country Code', how='left')

# on we used to determine on which column we are going to combine
```

```python
final_df.head(5)
```

## To check data types

```python
final_df.dtypes
```

```python
final_df.columns
```

```python
country_names = final_df.Country.value_counts().index
country_names
```

```python
country_val = final_df.Country.value_counts().values
country_val
```

## pie chart

```python
# pie chart - tp 3 countries that uses zomato
plt.pie (country_val[:3], labels = country_names[:3], autopct='%1.2f%%')
```

### Observation of above pie chart: Zomato maximum records or transaction are from india. After that USA and United kingdom. 

## Fetching Observations from data through making reltionships between different columns

```python
final_df.columns
```

```python
final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index()
# reset_index() for making dataframe
```

```python
# chaning column name
ratings= final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
ratings
```

#### Observation of ratings: 
 1. When rating is between 4.5 to 4.9 ----> Excellent
 2. When rating is between 4.0 to 4.4 ----> very good
 3. When rating is between 3.5 to 3.9 ----> good
 4. When rating is between 3.0 to 3.4 ----> average
 5. When rating is between 2.5 to 2.9 ----> average
 6. When rating is between 2.0 to 2.4 ----> poor

### Plotting of dataframe Ratings

```python
ratings.head()
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x='Aggregate rating', y='Rating Count', data=ratings)
```

```python
# mapping the colors according to column 'rating color'
plt.figure(figsize=(12,6))
sns.barplot(x='Aggregate rating', y='Rating Count', hue='Rating color',data=ratings, palette=['blue','red','orange','yellow','green','green'])
```

#### Observations from ratings after plotting:
1. Not rated count is very high
2. Maximum number of ratings are between 2.5 to 3.4

```python
### count plot
sns.countplot(x='Rating color', data=ratings, palette=['blue','red','orange','yellow','green','green'])
```

#### 1 record of white, 10 records of orange etc in ratings dataframe.
#### This counts is not basically rating count. It is the frequency of records present in dataset
